### 字段说明

字段|说明
-|-
IM | 房屋所在镇的犯罪率
ZN | 用地面积远大于 25000 平方英尺的住宅所占比例
INDUS | 房屋所在镇无零售业务区域所占比例
CHAS | 与查尔斯河有关的虚拟变量（如果房屋位于河边则值1，否则为0）
NOX | 一氧化碳浓度（每千万分之一）
RM | 每处寓所的平均房间数
AGE | 业主自住房屋中，建于1940年之前的房屋所占比例
DIS | 房屋举例波士顿五大就业中心的加权距离
RAD | 距离房屋最近公路入口编号
TAX | 每一万美元全额财产税金额
PTRATIO | 房屋所在镇的师生比
B | 计算公式为 1000(Bk - 0.63)^2，其中 Bk 为房屋所在镇非裔美籍人口所占比例
LSTAT | 弱势群体人口所占比例
MEDV | 业主自住房屋的平均价格（以1000美元为单位）

```python
# CRIM -> 房屋所在镇的犯罪率
# ZN -> 用地面积远大于 25000 平方英尺的住宅所占比例
# INDUS -> 房屋所在镇无零售业务区域所占比例
# CHAS -> 与查尔斯河有关的虚拟变量（如果房屋位于河边则值1，否则为0）
# NOX -> 一氧化碳浓度（每千万分之一）
# RM -> 每处寓所的平均房间数
# AGE -> 业主自住房屋中，建于1940年之前的房屋所占比例
# DIS -> 房屋举例波士顿五大就业中心的加权距离
# RAD -> 距离房屋最近公路入口编号
# TAX -> 每一万美元全额财产税金额
# PTRATIO -> 房屋所在镇的师生比
# B -> 计算公式为 1000(Bk - 0.63)^2，其中 Bk 为房屋所在镇非裔美籍人口所占比例
# LSTAT -> 弱势群体人口所占比例
# ---
# MEDV -> 业主自住房屋的平均价格（以1000美元为单位）
df.columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", 
              "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", 
              "LSTAT", "MEDV"]
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X, Y)
```

```python
print("Slope: %.3f" % slr.coef_[0])
print("Lntercept: %.3f" % slr.intercept_)
```

```python
def lin_regplot(X, Y, model):
    plt.scatter(X, Y, c="blue")
    plt.plot(X, model.predict(X), color="red")
    return None
```

```python
lin_regplot(X, Y, slr)
plt.xlabel("RM")
plt.ylabel("MEDV")
```

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),
                        max_trials=1000, # 最大迭代次数
                        min_samples=100, # 随机抽取作为内点的最小样本数量
                        residual_metric=lambda x: np.sum(np.abs(X), axis=1), # 计算拟合曲线与样本点间垂直距离的绝对值
                        residual_threshold=5.0, # 只有与拟合曲线垂直距离小于5个单位长度的样本点才能加入到内点集合
                        random_state=0)
ransac.fit(X, Y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_x = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_x[:, np.newaxis])
plt.scatter(X[inlier_mask], Y[inlier_mask], c="blue", marker="o", label="Inliers")
plt.scatter(X[outlier_mask], Y[outlier_mask], c="lightgreen", marker="s", label="Outliers")
plt.plot(line_x, line_y_ransac, color="red")
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.legend(loc="upper left")
```

```python
print("Slope: %.3f" % ransac.estimator_.coef_[0])
print("Lntercept: %.3f" % ransac.estimator_.intercept_)
```

```python
x = df[["LSTAT"]].values
y = df["MEDV"].values
x, y
```

```python
regr = LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
_4 = PolynomialFeatures(degree=4)
x_quad = quadratic.fit_transform(x)
x_cubic = cubic.fit_transform(x)
x_4 = _4.fit_transform(x)
x_quad, x_cubic, x_4
```

```python
from sklearn.metrics import r2_score
x_fit = np.arange(x.min(), x.max(), 1)[:, np.newaxis]
regr = regr.fit(x, y)
y_lin_fit = regr.predict(x_fit)
linear_r2 = r2_score(y, regr.predict(x))
linear_r2
```

```python
regr = regr.fit(x_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(x_fit))
quadratic_r2 = r2_score(y, regr.predict(x_quad))
quadratic_r2
```

```python
regr = regr.fit(x_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(x_fit))
cubic_r2 = r2_score(y, regr.predict(x_cubic))
cubic_r2
```

```python
regr = regr.fit(x_4, y)
_4_r2 = r2_score(y, regr.predict(x_4))
_4_r2
```

```python
def t(x, y, n):
    regr = LinearRegression()
    pf = PolynomialFeatures(degree=n)
    x_pf = pf.fit_transform(x)
    regr = regr.fit(x, y)
    regr = regr.fit(x_pf, y)
    return r2_score(y, regr.predict(x_pf))
```

```python
for n in range(1, 20):
    print(t(x, y, n))
```
